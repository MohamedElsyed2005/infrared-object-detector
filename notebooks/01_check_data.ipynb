```python
import yaml

with open("../data/test/data.yaml") as f:
    test_yaml = yaml.safe_load(f)
    
with open("../data/train/data.yaml") as f:
    train_yaml = yaml.safe_load(f)

with open("../data/val/data.yaml") as f:
    valid_yaml = yaml.safe_load(f)

print("Test classes:", test_yaml['names'])
print("Train classes:", train_yaml['names'])
print("Valid classes:", valid_yaml['names'])
print("Test nc:", test_yaml['nc'])
print("Train nc:", train_yaml['nc'])
print("Valid nc:", valid_yaml['nc'])
```
